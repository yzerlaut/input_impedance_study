### Quantitative analysis of the distortion introduced by the non-zero access resistance and non-infinite seal resistance in patch recordings

```python
import numpy as np
import matplotlib
import matplotlib.pylab as plt
```

```python
# impedance of the system if the cell is just a RC circuit
def Imped_at_amplifier(w, Rm, Cm, Rs, Ra):
    #return Ra+(1./Rs+(1+1j*w*Rm*Cm)/Rm)**(-1)
    return (1./Rs+(1.+1j*w*Rm*Cm)/(Ra*(1.+1j*w*Rm*Cm)+Rm))**(-1)


def plot_modulus_slope_and_phase(w, Z, axes=None, sliding_window=4, color='k'):  
    if axes is None:
        ax1=plt.subplot(121)
        ax3=plt.axes((.18,.22,0.1,0.2))
        ax2=plt.subplot(122)
    else:
        ax1, ax2, ax3 = axes
  
    ax1.loglog(w, 1e-6*np.abs(Z), color=color)
    ax2.semilogx(w, -np.angle(Z), color=color)
    ax2.set_ylabel('phase shift (Rd)')
    ax2.set_yticks([0, np.pi/4., np.pi/2.])
    ax2.set_yticklabels(["0","$\pi$/4","$\pi$/2"])

    # slope calculus
    x,y = [], []
    for i in range(sliding_window,len(w)-sliding_window):
        pol = np.polyfit(np.log(w[i-sliding_window:i+sliding_window]),
                         np.log(np.abs(Z[i-sliding_window:i+sliding_window])),1)
        x.append(w[i])
        y.append(pol[0])

    ax3.semilogx(x, y, color=color)
    ax3.set_ylabel('slope', size='small')
    ax3.set_yticks([0,-.5,-.75,-1])
    ax3.set_yticklabels(["0","-0.5","","-1"])
    ax3.grid()
    
    return [ax1, ax2, ax3]
```

```python
w = np.logspace(0,3,1e2)
Ra, Rs, Rm, Cm = 5e6, 10e9, 100e6, 200e-12

```

```python
Z = Imped_at_amplifier(w, Rm, Cm, Rs, Ra)
fig = plt.figure(figsize=(10,4))
plt.suptitle('The $\sim$ best you can get, $R_a$='+str(int(1e-6*Ra))+\
    'M$\Omega$, $R_s$='+str(int(1e-9*Rs))+'G$\Omega$    (with Single Comp. Hyp.)')
axes = plot_modulus_slope_and_phase(w, Z)
```

```python
w = np.logspace(0,3,1e2)
RS = np.logspace(0, 3, 10)*1e6 #
levels = 1e-6*RS

mymap = matplotlib.colors.LinearSegmentedColormap.from_list(\
            'mycolors',['red','blue'])
Z = [[0,0],[0,0]]
CS3 = plt.contourf(Z, levels, cmap=mymap)
plt.clf()

Ra, Rm, Cm = 5e6, 100e6, 200e-12

fig = plt.figure(figsize=(10,4))
plt.suptitle('Effect of the Seal resistance, here $R_a$='+str(int(1e-6*Ra))+'M$\Omega$')
axes=None
for i in range(len(levels)):
    r = (levels[i]-levels.min())/(levels.max()-levels.min())
    Z = Imped_at_amplifier(w, Rm, Cm, RS[i], Ra)
    
    axes = plot_modulus_slope_and_phase(w, Z, axes=axes, color=mymap(r,1))

ax1, ax2, ax3 = axes
cb = plt.colorbar(CS3, use_gridspec=True)
cb.set_label('Seal Resistance $R_S$ (M $\Omega$)')
```
